```python
import os
import datetime
import sys
sys.path.append('C:\Khue\H9_Solar_Power_Forecasting\src')
from image_processing import crop_and_resize_image


def batch_process_images(base_input_path, base_output_path, plant_name):
    """
    Láº·p qua cÃ¡c thÆ° má»¥c ngÃ y (20250101 - 20250314), thÆ° má»¥c giá» (0000 - 0550, cÃ¡ch 10 phÃºt)
    vÃ  xá»­ lÃ½ táº¥t cáº£ áº£nh trong má»—i thÆ° má»¥c.
    
    Parameters:
    - base_input_path (str): ThÆ° má»¥c gá»‘c chá»©a áº£nh Ä‘áº§u vÃ o.
    - base_output_path (str): ThÆ° má»¥c gá»‘c chá»©a áº£nh Ä‘áº§u ra.
    - plant_name (str): TÃªn nhÃ  mÃ¡y cáº§n xá»­ lÃ½.
    """
    
    # 1. Táº¡o danh sÃ¡ch cÃ¡c ngÃ y tá»« 20250101 Ä‘áº¿n 20250314
    start_date = datetime.date(2025, 3, 1)
    end_date = datetime.date(2025, 3, 14)
    date_list = [(start_date + datetime.timedelta(days=i)).strftime("%Y%m%d") 
                 for i in range((end_date - start_date).days + 1)]

    # 2. Láº·p qua tá»«ng ngÃ y
    for date_folder in date_list:
        date_path = os.path.join(base_input_path, date_folder)
        if not os.path.exists(date_path):
            continue  # Bá» qua náº¿u thÆ° má»¥c khÃ´ng tá»“n táº¡i

        print(f"ğŸ“‚ Äang xá»­ lÃ½ ngÃ y: {date_folder}")

        # 3. Láº·p qua táº¥t cáº£ thÆ° má»¥c giá» trong ngÃ y
        for time_folder in os.listdir(date_path):
                time_path = os.path.join(date_path, time_folder)

                if not os.path.exists(time_path):
                    continue  # Bá» qua náº¿u thÆ° má»¥c khÃ´ng tá»“n táº¡i

                print(f"  ğŸ•’ Äang xá»­ lÃ½ thÆ° má»¥c giá»: {time_folder}")

                # 4. Láº·p qua táº¥t cáº£ áº£nh trong thÆ° má»¥c
                for image_file in os.listdir(time_path):
                    # Kiá»ƒm tra náº¿u file lÃ  áº£nh vÃ  cÃ³ chá»©a cÃ¡c cá»¥m tá»« cáº§n thiáº¿t
                    if (image_file.lower().endswith(('.jpg', '.png', '.tif', '.jpeg')) and 
                        any(keyword in image_file.lower() for keyword in ['b03', 'b07', 'b08', 'b13','vir'])):
                        image_path = os.path.join(time_path, image_file)

                        try:
                            # Gá»i hÃ m xá»­ lÃ½ áº£nh
                            output_path = crop_and_resize_image(image_path, plant_name, base_input_path, base_output_path)
                            
                            if output_path:
                                print(f"    âœ… ÄÃ£ xá»­ lÃ½: {image_file} -> {output_path}")
                        except Exception as e:
                            print(f"    âš ï¸ KhÃ´ng thá»ƒ xá»­ lÃ½ áº£nh {image_file}: {str(e)}")
                        continue

    print("ğŸ‰ HoÃ n thÃ nh xá»­ lÃ½ táº¥t cáº£ áº£nh!")


```
